```python
import os
import shutil


def copy_game_dirs(df, source_dir, target_dir, column_name):
    game_ids = df[column_name].tolist()
    os.makedirs(target_dir, exist_ok=True)

    for game_id in game_ids:
        source_game_dir = os.path.join(source_dir, str(game_id))
        target_game_dir = os.path.join(target_dir, str(game_id))

        if os.path.exists(source_game_dir):
            shutil.copytree(source_game_dir, target_game_dir)
        else:
            print(f"Directory does not exist: {source_game_dir}")
```

```python
source_dir = '../data/Assets'
```

```python
target_dir = '../data/assets_with_large_ER'
copy_game_dirs(top_100_ER, source_dir, target_dir, 'game_id')
```

```python
target_dir = '../data/assets_with_large_CTR'
copy_game_dirs(top_100_CTR, source_dir, target_dir, 'game_id')
```

```python
import os
import shutil

def merge_game_dirs(game_ids, source_dir, target_dir):
    os.makedirs(target_dir, exist_ok=True)

    for game_id in game_ids:
        source_game_dir = os.path.join(source_dir, str(game_id))
        target_game_dir = os.path.join(target_dir, str(game_id))

        if os.path.exists(source_game_dir):
            shutil.copytree(source_game_dir, target_game_dir)
        else:
            print(f"Directory does not exist: {source_game_dir}")
```

```python
target_dir = '../data/assets_with_large_rate'
common_game_ids = set(top_100_ER['game_id']).intersection(top_100_CTR['game_id'])
merge_game_dirs(common_game_ids, source_dir, target_dir)
```
